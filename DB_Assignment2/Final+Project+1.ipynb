## Part 1: Scrape the Data
### We scrape some of the data from http://www.atpworldtour.com and some data from github.

```python
#Following is the code for scraping the coach data.
```

```python
from lxml import html
import requests
import re
import json
import csv
import sys
import time
import numbers
from concurrent.futures import ProcessPoolExecutor
import concurrent.futures

def html_parse_tree(url):
    page = requests.get(url)
    tree = html.fromstring(page.content)
    return tree

def xpath_parse(tree, xpath):
    result = tree.xpath(xpath)
    return result

def regex_strip_string(string):
    string = re.sub('\n', '', string).strip()
    string = re.sub('\r', '', string).strip()
    string = re.sub('\t', '', string).strip()
    return string

def regex_strip_array(array):
    for i in range(0, len(array)):
        array[i] = regex_strip_string(array[i]).strip()
    return array

def array2csv(array, filename):
    csv_array = array
    csv_out = open(filename + ".csv", 'w')
    mywriter = csv.writer(csv_out)
    for row in csv_array:
        mywriter.writerow(row)
    csv_out.close()
```

```python
import sys
coach_url = "http://www.atpworldtour.com/en/players/coaches"     
coach_tree = html_parse_tree(coach_url)

coach_name_xpath = '//td//a[@class="coach-name"]/text()'
coach_name_parsed = xpath_parse(coach_tree, coach_name_xpath)
coach_name_cleaned = regex_strip_array(coach_name_parsed) 

coach_count = len(coach_name_cleaned)

coach = []
for h in range(0, coach_count):
    coach_id = h+1
    coach_name = coach_name_cleaned[h]    
    try:
        coach_name_split = coach_name.split(', ')
        coach_firstname = coach_name_split[1]
        coach_lastname = coach_name_split[0]
    except Exception:
        coach_firstname = ''
        coach_lastname = ''
    coach_fullname = coach_firstname + '-' + coach_lastname 
    coach.append([coach_id, coach_fullname])    
filename = 'coach_fullname'
array2csv(coach, filename)  
print("success")
```
